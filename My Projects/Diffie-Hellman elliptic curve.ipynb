```python
def find_nth_derivative(expr,nth,center):
    return diff(expr,x,nth).evalf(subs = {x:center})


def taylor_coeffs(function,nth,center):
    result = []
    for n in range(nth):
      
        deriv = find_nth_derivative(function,n,center)
        # taylor_coefficient = (deriv/math.factorial(n))
        result.append(deriv)
    return result

def calc_polynomial(polynomial_coefficients, x_value,center):
    result = 0
    for index in range(len(polynomial_coefficients)):
        result+= (polynomial_coefficients[index] * ((x_value-center) ** index))/math.factorial(index)
    return result


def plot_taylor_series(function,derivatives_count,center,x_min=-1,x_max=1):
    
    x_axis = np.linspace(x_min,x_max,500)
    y_original = [find_nth_derivative(function,0,x) for x in x_axis]
    plt.title('Taylor series plots')
    plt.plot(x_axis, y_original, label = r'Original function')
    plt.scatter(center,find_nth_derivative(function,0,center))
    ax = plt.gca()
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.xlabel('x')
    plt.ylabel('y')

    for nth_derivative in range(derivatives_count):
        nth_dy = [calc_polynomial(taylor_coeffs(function,nth_derivative + 2,center), x, center) for x in x_axis]
        plt.plot(x_axis,nth_dy, label = f'{nth_derivative + 1} derivative')
    plt.legend()
    plt.show()
```

```python
import pygame
import numpy as np
def animation(screen_width = 1500,screen_height = 800):

    center = (screen_width / 2, screen_height / 2)
    
    pygame.init()
    win = pygame.display.set_mode((screen_width, screen_height))
    bg_color = (255, 255, 255)

    pygame.display.set_caption('Expanding spiral')
    
   
    angle_speed = np.pi/4
    font = pygame.font.Font(None, 36)

    clock = pygame.time.Clock()
    FPS = 100  # frames per second
    clock.tick(FPS)

    running = True
    frame = 0

    
        
    while running:
        frame += 1
        speed = frame/4
        # Check for event
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        win.fill(bg_color)
        # position = np.array(center)

        # Draw expanding cirlce
        pygame.draw.circle(win, (255, 165, 0), center, speed)
        pygame.draw.circle(win, (255, 255, 255), center, speed-1)
        
        # Draw axeses
        pygame.draw.line(win, (0,0,0), [750,0], [750,800], 1)
        pygame.draw.line(win, (0,0,0), [0,400], [1500,400], 1)        
        
        x_position = speed * np.cos(angle_speed * speed/10) + center[0]
        y_position = center[1] - speed * np.sin(angle_speed * speed/10) 
        pygame.draw.circle(win, (0, 0, 0), [x_position, y_position], 3)
        
        pygame.draw.line(win, (250,0,0),[x_position,center[1]], [center[0],y_position],2)
        pygame.draw.line(win, (250,0,0),[center[0],center[1]], [x_position,y_position],2)
        
        
        text = font.render(f'Frame number {frame}', True, (0, 0, 0))
        # Refreshing the screen
        win.blit(text, (10, 10))
        clock.tick(FPS)
        pygame.display.update()
        pygame.display.flip()
        
        if frame == 2000:
            running = False
       
    # Exit
    pygame.quit()
animation()
```

```python
def distance(point1, point2):
    return math.sqrt((point2[0] - point1[0]) ** 2 + (point2[1] - point1[1]) ** 2)
```

```python
def locate_spiral_radius(point_dist, angle_velocity, radius_velocity):
    # Find time `t`
    t = point_dist/ radius_velocity
    
    # Find angle of spiral rotation for a given time `t`
    theta = t * angle_velocity
    # Calcutate `x` and `y` coordinates
    x =  t * radius_velocity * np.cos(t * angle_velocity)
    y =  t * radius_velocity * np.sin(t * angle_velocity)    
    
    return np.array([x, y])
```

```python
 def slope(point1, point2):
        
    y_length = point1[1] - point2[1] 
    x_length = point1[0] - point2[0]
    if x_length == 0 and y_length == 0:
        return None
    if x_length == 0:
        return 'inf'
    elif y_length == 0:
        return 0
    return y_length / x_length
```

```python
def get_angle(slope, degrees = False):
    if slope == 'inf':
        if degrees:
            return 90
        return np.pi/2
    if degrees:
        return np.arctan(slope) * 180/ np.pi
    return np.arctan(slope)
    
```

```python
def get_radii(func):
    x_range = np.linspace(-20, 20, 1000)
    radii_list = []
    points = []
    for x in x_range:
        y = func(x)
        points.append([x, y])
        r = distance(np.array([x, y]), np.array([0, 0]))
        radii_list.append(r)
    return radii_list, points
```

```python
a_point = np.array([20,20])
def calc_first_intersection_point(point_of_linear_func, center_of_spiral_point, angle_velocity, initial_spiral_angle, radius_velocity):
    # Distance from the point of a function and the center of a spiral
    point_spiral_distance = distance(point_of_linear_func, center_of_spiral_point)
    print(f'Distance:  {point_spiral_distance}')
    
    # Slope of the line between the point and the center of spiral
    point_spiral_slope = slope(point_of_linear_func, center_of_spiral_point)
    print(f'Slope:  {point_spiral_slope}')
    
    # The angle of this slope in radians
    function_angle = get_angle(point_spiral_slope)
    
    # Recalculate the function angle according to its magnitude, positive or negative value and the direction of the spiral revolution
    if angle_velocity > 0 and function_angle < 0:
        function_angle = np.pi + function_angle
    if angle_velocity < 0 and function_angle >0:
        function_angle = - np.pi + function_angle
    print(f'Angle in radians:  {function_angle}')
    result_angle = function_angle + initial_spiral_angle
    print(result_angle)
    
    # Calculate the time corresponding to this value of the angle
    time = result_angle/ angle_velocity
    print(f'The time for this angle is: {time}')
    
    # The magnitude of the radius corresponding to this time
    curr_radius = radius_velocity * time
    
    print(f'Current radius = {curr_radius}')
    
    # Divide distance from point and curr radius
    ratio = point_spiral_distance / curr_radius
    print(f'Ratio bewtween distance and radius: {ratio}')
    
    # Full rotation time
    rotation_time = 2 * np.pi / angle_velocity
    
    print(f'Full rotation time: {rotation_time}')
calc_first_intersection_point(A_point, np.array([0 ,0]), angle_velocity, 0, radius_velocity)
```
