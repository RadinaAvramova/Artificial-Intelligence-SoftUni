```python
# Plot funcions
def show_hist(country_code):   
    
    country_name = world_population.loc[world_population['Country Code'] == country_code, 'Country Name'].iloc[0]
    
    years = world_population.loc[world_population['Country Code'] == country_code].drop(columns = ['Country Code', 'Country Name']).min().dropna().index
    
    min_year, max_year = years.min(), years.max()
    
    data = np.array(world_population.loc[world_population['Country Code'] == country_code].drop(columns=['Country Name', 'Country Code']))[0]
    if country_code == 'PSE':
        data = np.array(pd.Series(data).dropna())
        
    mean_population, max_population, min_population = np.mean(data), np.max(data), np.min(data)
    
    diff_percentage = 100 - (min_population / max_population) * 100
    
    plt.figure(figsize=(10, 6))
    
    max_height= plt.hist(data, bins=15, edgecolor='black')
    mean_x = np.mean(data)
    
    plt.title('Population Histogram of {} between {}-{}'.format(country_name, min_year, max_year))

    plt.xlabel('Population')
    plt.ylabel('Count')

    plt.text(mean_x, max_height[0].max(), f'Mean - {mean_population:.0f} people', fontsize=9, ha='center')
    plt.text(mean_x, max_height[0].max() - 0.4, f'Max - {max_population:.0f} people', fontsize=9, ha='center')
    plt.text(mean_x, max_height[0].max() - 0.8, f'Min - {min_population:.0f} people', fontsize=9, ha='center')
    plt.text(mean_x, max_height[0].max() - 1.2, f'Percentage - {diff_percentage:.2f}%', fontsize=9, ha='center')
    
    plt.show()
    
def plot(country_code):
    data = world_population[world_population['Country Code'] == country_code]
    years = data.columns[2:]
    min_year, max_year = np.min(years), np.max(years)
    population = data.iloc[0, 2:]
    country_name = world_population.loc[world_population['Country Code'] == country_code, 'Country Name'].iloc[0]
    if country_code == 'PSE':
        data = data.drop(columns=['Country Code', 'Country Name'])      
        data = data.dropna(axis = 1)
        years =  data.columns
        min_year, max_year = np.min(years), np.max(years)
        population = data.iloc[0]
        
    plt.figure(figsize=(17, 6))  
    plt.title(f'Population growth in {country_name} between {min_year} - {max_year}')
    plt.plot(years, population)
    plt.xticks(rotation = 45)
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.show()
```

# Study of population growth in several countries

## Bulgaria

```python
show_hist('BGR')
```

The histogram reveals a significantly uneven distribution with a considerable difference between the maximum and minimum population counts for the given period - around 28%. The dynamics of this change will be disclosed by the population plot over the years.

```python
plot('BGR')
```

There is significant movement in population growth over the specified period. Between 1960 and 1990, there is a pronounced growth, with its peak around 1990. Then, there is a sharp decline until 2022.


In the current dataset, there are no more data from which we can hypothesize about the possible reasons for these population growth movements. Therefore, we will turn to other sources. We need to examine the birth and death rates in Bulgaria for the period 1960-2022. Suitable data is available at this [address](https://en.wikipedia.org/wiki/Demographics_of_Bulgaria#Historical_birth_and_fertility_rates).

```python
# Load dataset
births_and_deaths = pd.read_csv("data/Demographics Bulgaria.csv", header=None)
```

```python
births_and_deaths
```

```python
# Leave only the necessary columns 
births_and_deaths = births_and_deaths.drop(columns=[1, 4, 5, 6, 7])
births_and_deaths.head()
```

```python
# Rename columns 
births_and_deaths = births_and_deaths.rename(columns = {0: 'year', 2: 'births', 3: 'deaths'})
births_and_deaths.head(20)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(births_and_deaths.year, births_and_deaths.births, label = 'Births')
plt.plot(births_and_deaths.year, births_and_deaths.deaths, label = 'Deaths')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

As seen in the mortality graph in Bulgaria from 1960 onwards, there is a noticeable upward trend. Just before 2000 and up to the year 2020, this trend is replaced by a very slight decrease in the number of deaths. Finally, in 2020, there is a very sharp increase followed immediately by a very sharp decline - this coincides with the COVID-19 pandemic.

The birth rate graph also shows an interesting pattern. From 1960 until slightly after the mid-1970s, the trend is increasing, although there are well-defined fluctuations up and down. However, this is followed by a declining trend, which can be divided into two parts - a less pronounced decline between the late 1970s and the late 1980s. Then comes a more pronounced decline from the late 1980s almost until the late 1990s. After that, there are minor fluctuations until 2022, and from 2010 onwards, a decline begins again.

The most interesting point in the graph is, of course, the intersection point right at the end of the 1980s. This coincides with one of the most significant events for the country in the 20th century - the collapse of the communist regime and the opening of the country to the West. On the previous graph, the same period is very clearly visible - it is the peak in the population growth of the country, when it reaches around 9 million people, followed by a strongly pronounced negative growth until it decreases by almost one-third by the end of 2022.

From 1990 onwards, a permanent trend of negative population growth begins in Bulgaria.

```python
# Calculating the absolute population growth for the periods from 1960 to 1989 and from 1990 to 2022.
positive_growth, negative_growth = 0, 0

for index, row in births_and_deaths.iterrows():
    population_growth = row.births - row.deaths
    if row.year < 1990:
        positive_growth += population_growth
    else:
        negative_growth += population_growth
positive_growth, negative_growth  
```

```python
print(f'Population 1960: {world_population[world_population.index == 21]["1960"].values[0]}')
print(f'Population 1960 + positive growth (1960-1989): {world_population[world_population.index == 21]["1960"].values[0] + positive_growth}')
print(f'Maximum population (1960-2022): {world_population[world_population.index == 21]["1988"].values[0]}')
```

We see that the positive population growth from 1960 to 1988-1989 can well explain the increase during this period.

The entire predicted population increase - according to the dynamics of births and deaths - is about 234,000 people more than the actual population count in 1988. This could be due to other factors, such as the migration of Bulgarians to Western European countries and the United States, which began at a slow pace during the communist regime.

```python
print(f'Population 1988: {world_population[world_population.index == 21]["1988"].values[0]}')
print(f'Population 1988 + nagative growth (1988-2022): {world_population[world_population.index == 21]["1988"].values[0] + negative_growth}')
print(f'Minimum population (1960-2022): {world_population[world_population.index == 21]["2022"].values[0]}')
```

Here the difference between the predicted and actual values is even greater - around 1,153,000. From 1990 onwards, with the opening of borders, a process of mass migration from the country began, which continues to this day. This appears to be the most likely reason for this discrepancy.
