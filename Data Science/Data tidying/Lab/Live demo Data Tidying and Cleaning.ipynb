```python
# Grouping data
mean_counts_by_religion = pew_tidy.groupby('religion').frequency.mean().sort_values()
```

```python
mean_counts_by_religion
```

```python
# Grouping returns a tuple with two values - the group and its values.
for key, value in pew_tidy.groupby('religion'):
    print(key, value.frequency.min(), value.frequency.mean(), value.frequency.max(), sep = ', ')
```

```python
plt.barh(mean_counts_by_religion.index, mean_counts_by_religion)
plt.xlabel('Mean number of peaple')
plt.title('Mean number of people bu religion')
plt.show()
```

```python
# Creating data frame
aggregates = {
    'religion' : [],
    'mean' : [],
    'min' : [],    
    'max' : [],
    
}
for key, value in pew_tidy.groupby('religion'):
    
    aggregates['religion'].append(key)
    aggregates['mean'].append(value.frequency.mean())
    aggregates['min'].append(value.frequency.min())
    aggregates['max'].append(value.frequency.max())
    
    
aggregates_tidy = pd.DataFrame(aggregates)
aggregates_tidy
```

```python
red_wines = pd.read_csv('Data exercice/winequality-red.csv')
```

```python
red_wines
```

```python
rows = []
for data in red_wines.values:
    new_data= data[0].replace(';', ',').split(',')
    new_data = list(map(float, new_data))
    rows.append(new_data)
    # print(new_data, len(new_data))
rows = np.array(rows)
rows[:,0]
```

```python
new_columns = {0: 'fixed acidity',1: 'volatile acidity',2:'citric acid',
                     3: 'residual sugar', 4: 'chlorides', 5: 'free sulfur dioxide', 
                     6: 'total sulfur dioxide', 7: 'density', 8: 'pH', 9: 'sulphates',
                     10: 'alcohol', 11:'quality'}
new_columns
```

```python
for key, value in new_columns.items():
    print(key, value)
    red_wines[value] = rows[:,key]
    
```

```python
red_wines 
```

```python
red_wines = red_wines.drop(columns = ['fixed acidity;"volatile acidity";"citric acid";"residual sugar";"chlorides";"free sulfur dioxide";"total sulfur dioxide";"density";"pH";"sulphates";"alcohol";"quality"'])
```

```python
red_wines_quality = red_wines.quality.mean()
```

```python
red_wines_alcohol =  red_wines.alcohol.mean()
red_wines_alcohol
```

```python
round(red_wines_alcohol/red_wines_quality,2)
```
