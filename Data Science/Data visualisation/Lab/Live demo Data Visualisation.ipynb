```python
# Plot two datasets 
plt.hist(report_2018['GDP per capita'], label = '2018', alpha = 0.7)
plt.hist(report_2019['GDP per capita'], label = '2019', alpha = 0.7)
plt.legend()
plt.xlabel('GDP contribution')
plt.ylabel('Count')
plt.show()
```

```python
# Plot two datasets for score
plt.hist(report_2018['Score'], label = '2018', alpha = 0.7)
plt.hist(report_2019['Score'], label = '2019', alpha = 0.7)
plt.legend()
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot
plt.boxplot([report_2018['Score'],report_2019['Score']])
plt.xticks((1,2), ['2018', '2019'])
plt.show()
```

```python
# Merging two datasets
report_2018.merge(report_2019, left_on = 'Country or region', right_on = 'Country or region')
```

```python
# Replace x and y suffixes
merged_reports = report_2018.merge(report_2019, left_on = 'Country or region', right_on = 'Country or region', suffixes = ['_2018', '_2019'] )
```

```python
merged_reports
```

```python
scores = merged_reports[['Country or region', 'Score_2018', 'Score_2019']]
```

```python
scores
```

```python
scores.nlargest(20, 'Score_2018')
```

```python
# Bar charts are used to plot categorical variables
plt.barh(scores.nlargest(20, 'Score_2018')['Country or region'], scores.nlargest(20, 'Score_2018')['Score_2018'])
plt.xlim(0, 8)
plt.gca().invert_yaxis()
plt.show()
```

```python
# Bar charts are used to plot categorical variables
plt.barh(scores.nsmallest(20, 'Score_2018')['Country or region'], scores.nsmallest(20, 'Score_2018')['Score_2018'])
plt.xlim(0, 8)
plt.show()
```

```python
# Set the size of the plot - tuple values are measured in inches
plt.figure(figsize=(20,20))
plt.barh(scores.nsmallest(20, 'Score_2018')['Country or region'], scores.nsmallest(20, 'Score_2018')['Score_2018'])
plt.xlim(0, 8)
plt.show()
```

```python
# Pie chart is also used for categorial variables
plt.pie([1, 5, 8], labels = ['a', 'b', 'c'])
plt.show()
```

```python
merged_reports = merged_reports.merge(report_2017.rename(columns = lambda col: col + '_2017'), left_on = 'Country or region', right_on = 'Country_2017', suffixes=['', ' 2017'])
```

```python
merged_reports
```

```python
merged_reports = merged_reports.rename(columns = {'Happiness.Score_2017': 'Score_2017'})
```

```python
merged_reports
```

```python
scores = merged_reports[['Country or region', 'Score_2017', 'Score_2018', 'Score_2019']]
```

```python
scores
```

```python
# Matplotlib cannot plot rows. This can be fixed by transpose.Values takes only values, ravel makes values one-dimensional array.
bg_scores = scores[scores['Country or region'] == 'Bulgaria'].T.drop('Country or region').values.ravel()
romania_scores = scores[scores['Country or region'] == 'Romania'].T.drop('Country or region').values.ravel()
```

```python
bg_scores
```

```python
years = list(range(2017,2020))
```

```python
plt.plot(years, bg_scores, label = 'Bulgaria', linewidth = 2, linestyle = 'dashed')
plt.scatter(years, bg_scores, marker= 's', s = 50)

plt.plot(years, romania_scores, label = 'Romania', linewidth = 2, linestyle = 'dashed')
plt.scatter(years, romania_scores, marker= 's', s = 50)

plt.legend(loc = 'lower right')
plt.xticks(years)
plt.title('Bulgaria happiness score')
plt.xlabel('Year')
plt.ylabel('Score')
plt.show()
```
